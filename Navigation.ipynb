```python
#develop graph
graph = {'A': ['A1', 'B'],
         'A1': ['A'],
         'B': ['A', 'B1', 'C'],
         'B1': ['B'],
         'C': ['B', 'C1', 'D'],
         'C1': ['C'],
         'D': ['C', 'D1'],
         'D1': ['D']}
```

```python
#find the shortest path
def find_shortest_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path
    shortest = None
    for node in graph[start]:
        if node not in path:
            newpath = find_shortest_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest
```
